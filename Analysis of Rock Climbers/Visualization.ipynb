```python
import sqlite3
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
data = pd.read_csv('female_boulderers.csv', low_memory=False)
data
```

```python
# clean data
data.drop(data[data['age'] >= 2022].index, inplace = True)
data.drop(data[data['climbing length'] >= 2022].index, inplace = True)
data = data.dropna()
data
```

Next step is to visualize data by each row, only get rid of nan value of the column, to get more data.

```python
data.to_csv('female.csv')
```

```python
x = data.loc[:, 'height (cm)']
y = data.loc[:,'score']

plt.scatter(x,y)
plt.title('Female Height and Grade average')
plt.show()
```

```python
import seaborn as sns
import numpy as np
from sklearn import datasets

db = pd.DataFrame.from_records(data, columns=['score','height (cm)','climbing length', 'age'])

#pd.plotting.scatter_matrix(db) 
sns.pairplot(db) 
plt.show()
```
